```python
home_values = pd.read_csv('./Data/median_home_values.csv')
```

```python
#the zip is actually the ZCTA5 which is usually the same as the zip but not always
```

```python
home_values.head()
```

```python
home_values = home_values[['GEO.display-label','HD01_VD01']]
```

```python
home_values['zcta'] = home_values['GEO.display-label']
home_values['median_home_value'] = home_values['HD01_VD01']
```

```python
home_values = home_values.drop(columns = ['GEO.display-label', 'HD01_VD01'])
```

```python
home_values = home_values.drop(0, axis = 0)
```

```python
home_values
```

```python
home_values['zcta'] =  home_values['zcta'].apply(lambda x: (x[-5:]))
```

```python
home_values.info()
```

```python
home_values['median_home_value'] = home_values['median_home_value'].apply(lambda x: np.NaN if x == '-' else str(x).replace(",", ''))
```

```python
home_values['zcta'] = home_values['zcta'].astype('int64')
```

```python
home_values['median_home_value'] = home_values['median_home_value'].astype('float32')
```

```python
home_values
```

```python
home_values.to_csv('./Data/clean_home_values.csv', index = False)
```

```python
edu = pd.read_csv('./Data/ACS_17_5YR_S1501_with_ann.csv')
```

```python
edu.columns

for c in edu.columns:
    print(c)
```

```python
edu = edu[["Geography", 'Percent; Estimate; Percent high school graduate or higher',
     "Percent; Estimate; Percent bachelor's degree or higher" ]]
```

```python
edu['Geography'] = edu['Geography'].apply(lambda x: (x[-5:]))
```

```python
edu.rename(columns={"Geography": "zcat", "Percent; Estimate; Percent high school graduate or higher": "percent_highschool_or_higher",
                   "Percent; Estimate; Percent bachelor's degree or higher": "percent_bachelors_or_higher"}, inplace=True)
```

```python
edu
```

```python
edu['zcta'] = edu['zcat']
```

```python
edu = edu.drop(columns = "zcat")
```

```python
edu.zcat = edu.zcat.astype("int64")
```

```python
edu['percent_highschool_or_higher'] = edu['percent_highschool_or_higher'].apply(lambda x: np.NaN if x == '-' else x)
```

```python
edu['percent_bachelors_or_higher'] = edu['percent_bachelors_or_higher'].apply(lambda x: np.NaN if x == '-' else x)
```

```python
edu[['percent_highschool_or_higher', 'percent_bachelors_or_higher']] = edu[['percent_highschool_or_higher', 'percent_bachelors_or_higher']].astype("float64")
```

```python
edu.isna().sum()
```

```python
edu.info()
```

```python
edu.to_csv("./Data/edu_clean.csv", index = False)
```

```python
poverty = pd.read_csv("./Data/ACS_17_5YR_DP03_with_ann.csv")
```

```python
poverty.head()
```

```python
for c in poverty.columns:
    print(c)
```

```python
poverty = poverty[["Geography", "Estimate; COMMUTING TO WORK - Mean travel time to work (minutes)",
         "Percent; HEALTH INSURANCE COVERAGE - Civilian noninstitutionalized population - No health insurance coverage",
        "Percent; PERCENTAGE OF FAMILIES AND PEOPLE WHOSE INCOME IN THE PAST 12 MONTHS IS BELOW THE POVERTY LEVEL - All people - 18 years and over - 65 years and over"]]
```

```python
poverty.rename(columns = {"Geography": "zcta",
                        "Estimate; COMMUTING TO WORK - Mean travel time to work (minutes)": "mean_commute_time",
                          "Percent; HEALTH INSURANCE COVERAGE - Civilian noninstitutionalized population - No health insurance coverage": "percent_no_health_insurance", 
                          "Percent; PERCENTAGE OF FAMILIES AND PEOPLE WHOSE INCOME IN THE PAST 12 MONTHS IS BELOW THE POVERTY LEVEL - All people - 18 years and over - 65 years and over": "percent_below_poverty"}, 
              inplace = True
            )
```

```python
poverty.info()
```

```python
poverty['zcta'] = poverty['zcta'].apply(lambda x: x[-5:]).astype("int64")
```

```python
poverty["mean_commute_time"] = poverty["mean_commute_time"].apply(lambda x: np.NaN if x == "-" else x)
```

```python
poverty["percent_below_poverty"] = poverty["percent_below_poverty"].apply(lambda x: np.NaN if x == "-" else x)
```

```python
poverty["percent_no_health_insurance"] = poverty["percent_no_health_insurance"].apply(lambda x: np.NaN if x == "-" else x)
```

```python
poverty["mean_commute_time"].value_counts()
```

```python
poverty[["mean_commute_time", "percent_no_health_insurance", "percent_below_poverty" ]] = poverty[["mean_commute_time", "percent_no_health_insurance", "percent_below_poverty" ]].astype("float64")
```

```python
poverty.to_csv('./Data/poverty.csv', index = False)
```
